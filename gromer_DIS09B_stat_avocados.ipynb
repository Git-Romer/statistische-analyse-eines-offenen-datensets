Die haupt Fragestellung der Analyse wird sein:  
- **Gibt es bei der Verpackung der Avocados einen Unterschied zwischen ökologisch und konventionell?**

Die dazu passenden Hypothesen sind:  
- H<sub>0</sub>: **Ökologische Avocados werden eher in kleineren Säcken verkauft.**

- Hypothese H<sub>0</sub>: $\mu=70\%$ &rarr; Anteil verkaufter kleiner Säcke (Ökologisch)  
- Abweichung: $\alpha=0.05$

## Visualisierungen und Analyse

### Allgemeines Verständnis des Datensatzes

Um nicht sofort von vielen Grafen überrumpelt zu werden, fange ich an, mir einen Überblick zu verschaffen.

```python
import pandas as pd
import math
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt
dataset = pd.read_csv("./avocado.csv")
dataset
```

Anhand der Angezeigten Tabelle kann ich nun entnehmen, dass es sich um 18249 reihen und 14 Spalten handelt.  
Da die Tabelle sauber scheint, möchte ich nach Daten schauen, welche meine Auswertung vereinfachen könnten.  
Die spezifischen Regionen haben keine Relevanz für meine Statistik, weswegen ich vor habe, alle Regionen zu addieren und einen Gesamtwert zu bilden.
Dafür schaue ich mir die Spalte `region` an.

```python
print(set(dataset["region"]))
```

Sofort fällt auf, dass ich eine reihe von Regionen habe, aus denen die Daten stammen. Hier sticht die Region `TotalUS` heraus. Diese Region könnte bereits der Gesamtwert sein, welchen ich berechnen wollte.  
Um dies zu überprüfen, wird die Summe aller verkauften Säcke (`Total Bags`) in der Region `TotalUS` zu jedem Datum (`Date`) mit der gleichen Summe in allen Regionen ausschließlich der Region `TotalUS` verglichen.

```python
# Indexierung aller Daten in der Spalte "Date"
date = set(dataset["Date"])
vector = []
for dates in date:
    # Daten für TotalUS
    cur_date = dataset["Date"] == dates
    TotalUS_region = dataset["region"] == "TotalUS"
    TotalUS_data = dataset[TotalUS_region & cur_date]
    TotalUS_sum = TotalUS_data["Total Bags"].sum()
    # Daten für alles außer TotalUS
    not_TotalUS_region = dataset["region"] != "TotalUS"
    not_TotalUS_data = dataset[not_TotalUS_region & cur_date]
    not_TotalUS_sum = TotalUS_data["Total Bags"].sum()
    # Überprüfung ob beide Summen gleich sind
    vector.append(TotalUS_sum == not_TotalUS_sum)
# Ausgabe
print(*set(vector))
```

Im Code wurden beide Summen zusammengerechnet und ein `True` ausgegeben, falls beide Summen am selben Datum gleich waren. Duch die Ausgabe `True` bestätigt sich somit meine Annahme.

### Mathematische Berechnung der Hypothese

Nachdem ich meine Daten genug visualisiert habe, und auch weiß, welche Zahlen ich brauche, kann ich zu dem eigentlichen Teil meiner Statistik kommen.  
Die allgemeine Formel für einen t-Test lautet wie folgt:  
$\overline{x}-\mu \over \sqrt{S^{*2}/n}$

In diesem Sinne werde ich Schritt für Schritt alle Werte einsetzen, um die Hypothese zu belegen.

___

#### Schritt 1
$\overline{x}-\mu$ = (Durchschnitt aller `percentage_SB` der ökologischen Avocados) - 75

```python
# Festgelegter mü-Wert
mü = 0.70
# Mittelwert
xq = sum(only_organic["percentage_SB"]) / len(only_organic["percentage_SB"])
# Berechnung von Formel Schritt 1
res = xq - (mü * 100)
print(res)
```

___

#### Schritt 2

$S^{*2} = \frac{1}{n}\sum_{i = 1}^{n}(x_i-\overline{x})^2$

```python
# Berechnung von Formel Schritt 2
var_res = sum([(xi - xq) ** 2 for xi in only_organic["percentage_SB"]]) / len(only_organic["percentage_SB"])
print(var_res)
```

___

#### Schritt 3

$\sqrt{S^{*2}/n}$

```python
# Berechnung von Formel Schritt 3
math.sqrt(var_res/(len(only_organic["percentage_SB"])))
```

___

#### Schritt 4

$\overline{x}-\mu \over \sqrt{S^{*2}/n}$

```python
# Berechnung von Formel Schritt 4
(res)/(math.sqrt(var_res/(len(only_organic["percentage_SB"]))))
```

### Lösung

Nun kennen wir unsere Teststatistik $t=-1.668461031745599$  
Diesen Wert kann ich nun mit der Tabelle abgleichen, um herauszufinden, ob ich mit meiner Nullhypothese richtig lag.  
Dafür nutze ich den zweiseitigen Test bei einem bereits festgelegtem Signifikanzniveau.
